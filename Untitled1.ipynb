```python
import pandas as pd

# Load the dataset
file_path = '/content/Online Retail.xlsx'
data = pd.read_excel(file_path)
```

```python
data_cleaned = data.dropna(subset=['Description'])
```

```python
data_cleaned = data_cleaned.drop_duplicates()
data_cleaned
```

```python
data_cleaned['InvoiceDate'] = pd.to_datetime(data_cleaned['InvoiceDate'])
data_cleaned
```

```python
# Save the new data after clean to a new Excel file
data_cleaned.to_excel('Online_Retail_Cleaned.xlsx', index=False)

```

```python
#Feature Engineering

customer_features = data_cleaned.groupby('CustomerID').agg({
    'InvoiceNo': pd.Series.nunique,
    'Quantity': 'sum',
    'TotalPrice': 'sum',
    'InvoiceDate': ['max', 'min', 'nunique']
}).reset_index()

customer_features.columns = ['CustomerID', 'OrderFrequency', 'TotalQuantity', 'TotalSpending', 'LastPurchaseDate', 'FirstPurchaseDate', 'NumberOfPurchaseDays']

current_date = data_cleaned['InvoiceDate'].max()
customer_features['Recency'] = (current_date - customer_features['LastPurchaseDate']).dt.days

customer_features['AvgBasketSize'] = customer_features['TotalQuantity'] / customer_features['OrderFrequency']
customer_features['AvgSpendingPerOrder'] = customer_features['TotalSpending'] / customer_features['OrderFrequency']

customer_features = customer_features.drop(columns=['LastPurchaseDate', 'FirstPurchaseDate', 'NumberOfPurchaseDays'])
```

```python
customer_features
```

```python
#Dimensionality Reduction with PCA

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

features = customer_features.drop(columns=['CustomerID'])
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)

pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
pca_df['CustomerID'] = customer_features['CustomerID']
```

```python
#Determining Optimal Number of Clusters

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# Elbow
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(pca_result)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Silhouette analysis
silhouette_scores = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(pca_result)
    silhouette_avg = silhouette_score(pca_result, cluster_labels)
    silhouette_scores.append(silhouette_avg)

plt.plot(range(2, 11), silhouette_scores)
plt.title('Silhouette Analysis')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
#K-means Clustering

kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(pca_result)

# here we Add cluster labels to the DataFrame
pca_df['Cluster'] = clusters
```

```python
#Cluster Profiling

clustered_data = customer_features.merge(pca_df[['CustomerID', 'Cluster']], on='CustomerID')

cluster_profile = clustered_data.groupby('Cluster').agg({
    'OrderFrequency': 'mean',
    'TotalQuantity': 'mean',
    'TotalSpending': 'mean',
    'Recency': 'mean',
    'AvgBasketSize': 'mean',
    'AvgSpendingPerOrder': 'mean'
}).reset_index()

print(cluster_profile)
```

```python
#Visualization
import seaborn as sns

plt.figure(figsize=(10, 7))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=pca_df, palette='viridis')
plt.title('Customer categories')
plt.show()
```

```python
#Evaluation
silhouette_avg = silhouette_score(pca_result, clusters)
print(f'Silhouette Score: {silhouette_avg}')
```
