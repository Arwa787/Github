```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Step 2: Detecting Outliers

# For numerical columns, use the IQR method to detect outliers
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Make sure to use the original DataFrame 'data' for outlier detection
# Extract numerical columns from the 'data' DataFrame
numerical_columns = data.select_dtypes(include=[np.number]).columns.tolist()

outliers_dict = {}
for column in numerical_columns:
    outliers = detect_outliers_iqr(data, column) # Pass 'data' to the function
    if not outliers.empty:
        outliers_dict[column] = outliers

print("Outliers detected in numerical columns:")
for column, outliers in outliers_dict.items():
    print(f"Column: {column}, Number of outliers: {len(outliers)}")

# Visualize outliers using box plots
for column in numerical_columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(data[column]) # Use 'data' here as well
    plt.title(f"Box plot of {column}")
    plt.show()

# Handling Outliers: You can choose to remove or cap outliers
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Create a copy of the DataFrame to avoid modifying the original
data_cleaned = data.copy()
for column in numerical_columns:
    data_cleaned = remove_outliers_iqr(data_cleaned, column)

# Step 3: Handling Inconsistencies
# Example: Ensure all categorical values are consistent
# Select object type columns from 'data_cleaned'
categorical_columns = data_cleaned.select_dtypes(include=['object']).columns
for column in categorical_columns:
    unique_values = data_cleaned[column].unique()
    print(f"Unique values in {column}: {unique_values}")

# If any inconsistencies are found, you can handle them here
# For instance, if there are different cases for the same category
# Check if 'job' and 'marital' columns exist before applying lower()
if 'job' in data_cleaned.columns:
    data_cleaned['job'] = data_cleaned['job'].str.lower()
if 'marital' in data_cleaned.columns:
    data_cleaned['marital'] = data_cleaned['marital'].str.lower()

# Display the cleaned dataset
print("Cleaned dataset:")
print(data_cleaned.head())

# Save the cleaned dataset
data_cleaned.to_csv('cleaned_bank_additional_full.csv', index=False) # Remove '/mnt/data/' from the path
```
