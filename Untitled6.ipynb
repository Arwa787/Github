```python
import pandas as pd
from pycaret.classification import setup as cls_setup, compare_models as cls_compare_models, pull as cls_pull
from pycaret.regression import setup as reg_setup, compare_models as reg_compare_models, pull as reg_pull
import pycaret.classification
import pycaret.regression

# Load dataset
file_path = input("Enter the path of the dataset file (CSV or Excel): ")
if file_path.endswith('.csv'):
    df = pd.read_csv(file_path) # Assign the DataFrame to df here
elif file_path.endswith('.xlsx'):
    df = pd.read_excel(file_path) # Assign the DataFrame to df here

print("Here is a preview of your dataset:")
print(df.head())

# Handling missing values for continuous columns
cont_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
print("Continuous columns: ", cont_cols)
missing_handling = input("How would you like to handle missing values in continuous columns? (mean/median/mode): ")

if missing_handling.lower() == "mean":
    df[cont_cols] = df[cont_cols].fillna(df[cont_cols].mean())
elif missing_handling.lower() == "median":
    df[cont_cols] = df[cont_cols].fillna(df[cont_cols].median())
elif missing_handling.lower() == "mode":
    df[cont_cols] = df[cont_cols].fillna(df[cont_cols].mode().iloc[0])

print("Missing values handled.")

# Encoding categorical data
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
print("Categorical columns: ", cat_cols)
encoding_method = input("How would you like to encode categorical data? (one hot/label): ")

if encoding_method.lower() == "one hot":
    df = pd.get_dummies(df, columns=cat_cols)
elif encoding_method.lower() == "label":
    from sklearn.preprocessing import LabelEncoder
    le = LabelEncoder()
    for col in cat_cols:
        df[col] = le.fit_transform(df[col])

print("Categorical data encoded.")

# Choosing X and y
print("Columns in the dataset: ", df.columns.tolist())
target_col = input("Choose the target column: ")
feature_cols = input("Choose feature columns (comma separated): ").split(',')
feature_cols = [col.strip("'") for col in feature_cols]

X = df[feature_cols]
y = df[target_col]

# Detect task type
if df[target_col].dtype == 'object' or df[target_col].nunique() < 20:
    task_type = 'classification'
    print("Task Type: Classification")
else:
    task_type = 'regression'
    print("Task Type: Regression")

# Use PyCaret to train multiple models
print("Training Models with PyCaret")
if task_type == 'classification':
    cls_setup(data=df, target=target_col, silent=True, verbose=False) #Removed extra indentation
    best_model = cls_compare_models()
    report = cls_pull()
else:
    reg_setup(data=df, target=target_col, silent=True, verbose=False)
    best_model = reg_compare_models()
    report = reg_pull()

print("Model training completed and report generated.")
print(report)
```
