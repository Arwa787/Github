```python
from pycaret.datasets import get_data
from pycaret.classification import *
from pycaret.regression import *
import pandas as pd
import streamlit as st
```

```python
def main():
    st.title("AutoML with PyCaret and Streamlit")

    st.sidebar.title("Options")

    # Upload data
    data_file = st.sidebar.file_uploader("Upload your dataset (CSV file)", type=["csv"])

    if data_file is not None:
        data = pd.read_csv(data_file)
        st.write("Dataset:")
        st.write(data)

        # Select target variable
        target = st.sidebar.selectbox("Select the target variable", data.columns)

        # Identify problem type: classification or regression
        if data[target].dtype == 'object' or len(data[target].unique()) <= 20:
            problem_type = 'classification'
        else:
            problem_type = 'regression'

        st.write(f"Detected problem type: {problem_type}")

        if problem_type == 'classification':
            run_classification(data, target)
        else:
            run_regression(data, target)

def run_classification(data, target):
    if st.sidebar.button("Run Classification"):
        # Setup PyCaret classification
        clf_setup = setup_classification(data, target=target, silent=True, verbose=False)

        # Train and compare models
        best_model = compare_models_classification()
        st.write("Best Model:")
        st.write(best_model)

        # Display model comparison
        model_comparison = pull_classification()
        st.write("Model Comparison:")
        st.write(model_comparison)

def run_regression(data, target):
    if st.sidebar.button("Run Regression"):
        # Setup PyCaret regression
        reg_setup = setup_regression(data, target=target, silent=True, verbose=False)

        # Train and compare models
        best_model = compare_models_regression()
        st.write("Best Model:")
        st.write(best_model)

        # Display model comparison
        model_comparison = pull_regression()
        st.write("Model Comparison:")
        st.write(model_comparison)

if __name__ == '__main__':
    main()
```
